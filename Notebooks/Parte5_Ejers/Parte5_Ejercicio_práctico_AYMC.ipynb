## Ejercicio 13
Escribe una función que compruebe si un DNI es válido, devolviendo `True` en caso afirmativo.  
En primer lugar, el DNI debe tener entre 7 y 8 dígitos numéricos.  
Para saber si la letra del DNI es correcta puedes utilizar el siguiente fragmento de código:
>```Python
>DNI= 55555555
>palabra='TRWAGMYFPDXBNJZSQVHLCKE'
>print('La letra del DNI es: ', palabra[DNI%23])
>````


`````python
def dni_valido(dni):
  if 7 <= len(dni) <= 8:
    return True
  else:
    return False

def letra_dni_validar(dni):
  palabra='TRWAGMYFPDXBNJZSQVHLCKE'
  return palabra[int(dni)%23]

dni = input("Ingrese su DNI con letra: ")
if dni_valido(dni[:-1]):
  if letra_dni_validar(dni[:-1]) == dni[-1].upper():
    print("El DNI es correcta")
  else:
    print("La letra del DNI",dni,"no es correcta")
else:
  print("El DNI",dni,"no es válido")

`````
